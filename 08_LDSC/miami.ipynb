```python
import gwaslab as gl
```

```python
mysumstats = gl.Sumstats("BBJ_HDLC.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             p="P")
```

```python
mysumstats.basic_check()
```

```python
mysumstats.data.to_csv("BBJ_HDLC_miami.txt.gz",sep="\t",index=False)
```

```python
mysumstats = gl.Sumstats("BBJ_LDLC.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             p="P")
```

```python
mysumstats.basic_check()
```

```python
mysumstats.data.to_csv("BBJ_LDLC_miami.txt.gz",sep="\t",index=False)
```

```python
gl.plot_miami( 
          "BBJ_HDLC_miami.txt.gz",
          "BBJ_LDLC_miami.txt.gz",
          cols1=["CHR","POS","P"],
          cols2=["CHR","POS","P"],
          sep=["\t","\t"],
          anno= True
          )
```
