```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python

fig, ax = plt.subplots(figsize=(15,5),dpi=200)
x_0 = np.arange(-5, 10, 0.01)
# Calculating mean and standard deviation
mean_0 = 0
mean_1 = 3
sd = 1
  
ax.plot(x_0, norm.pdf(x_0, mean_0, 1),c="black")
ax.plot(x_0, norm.pdf(x_0, mean_1, 1),c="black")

x_beta = np.arange(-5, 2, 0.01)
x_alpha = np.arange(2, 10, 0.01)
ax.axvline(2,color="grey",linestyle="--")
x_power = np.arange(2, 8, 0.01)
ax.fill_between(x_beta, 0 , norm.pdf(x_beta, mean_1, 1)  , color="#EF0C01",label=r'$\beta$')
ax.fill_between(x_alpha, 0 ,norm.pdf(x_alpha, mean_0, 1) ,  fc="None", hatch='///',edgecolor='black' ,label=r'$\alpha$',zorder=100)
ax.fill_between(x_power, 0 ,norm.pdf(x_power, mean_1, 1) ,color="#2195F4",label=r'$Power = 1 - \beta$')
ax.text(-2, 0.3, r"$H_0$",fontsize=20)
ax.text(5, 0.3, r"$H_1$",fontsize=20)
ax.set_xlabel("Test Statistics",fontsize=20)
ax.set_ylabel("Frequency",fontsize=20 )
ax.legend()
```

```python
import scipy.stats as ss


f=0.5
b_alt=0.2
sigma = np.sqrt(1 -2 *f*(1-f)*(b_alt**2))
ns = np.arange(500, 4000, 10)
ses = sigma/ np.sqrt(ns*2*f*(1-f))
q_t = ss.chi2.ppf(1-5e-8,df=1)
power = 1 -ss.ncx2.cdf(q_t,nc=(b_alt/ses)**2,df=1)

fig, ax = plt.subplots(figsize=(15,5),dpi=200)
ax.plot(ns, power,c="black")
```
