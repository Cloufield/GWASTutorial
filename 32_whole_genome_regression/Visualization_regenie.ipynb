# Visualization

## Import gwaslab package

```python
import gwaslab as gl
```

## Load sumstats 

```python
sumstats = gl.Sumstats("1kg_eas_step1_BT_B1.regenie",fmt="regenie",sep=" ")
```

```python
sumstats.data
```

## Check the lead variants in significant loci

```python
sumstats.get_lead(sig_level=5e-6)
```

# Fill P

```python
sumstats.fill_data(to_fill=["P"])
```

## Create mahattan plot

```python
sumstats.plot_mqq(skip=2,anno=True)
```

## QC check

```python
sumstats.basic_check()
```

## Create regional plot

```python
sumstats.plot_mqq(mode="r",anno=True,region=(2,54531536,56731536),region_grid=True)
#2:
```

## Create regional plot with LD information

```python
sumstats.plot_mqq(mode="r",anno=True,region=(2,54531536,56731536),region_grid=True,vcf_path=gl.get_path("1kg_eas_hg19"))
```
